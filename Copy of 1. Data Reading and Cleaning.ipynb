# Change data type

Convert rain, soil, gwl and temperature data type from object to float.

```python
df['rain'] = pd.to_numeric(df['rain'], errors='coerce')
df['soil'] = pd.to_numeric(df['soil'], errors='coerce')
df['gwl'] = pd.to_numeric(df['gwl'], errors='coerce')
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
```

# Remove duplicates

Duplicate data with same datetime with same code and machine will be removed

```python
names = list(df['location'].unique())
names
```

```python
for i in names:
    raw_data = pd.read_pickle('raw/' + i + '.pkl')
    raw_data = raw_data.drop_duplicates(subset=['date'], keep='last')
    raw_data = raw_data.reset_index(drop=True)
    raw_data = raw_data.set_index('date')
    raw_data = raw_data.asfreq('10T', fill_value=np.nan)
    raw_data.reset_index(inplace=True)
    raw_data['location'] = i
    
    raw_data.to_pickle("date/" + i + ".pkl")
```
