## Read Data

```python
date_parser = lambda x: datetime.strptime(x, '%y%m%d%H%M%S')
df = pd.read_csv("data/sumsel_1Jam_Desember2022.csv", parse_dates = ['date'], date_parser = date_parser, sep=';')
df.head()
```

```python
karangagung = df[df["stasiun"] == "Karang agung"]
karangagung
```

```python
plt.plot(karangagung["date"], karangagung["gwl"])
```

```python
karangagung["date"].duplicated().sum()
```

```python
df1 = karangagung[~karangagung["date"].duplicated(keep='last')]
df1
```

```python
plt.plot(df1["date"], df1["gwl"])
```

```python
plt.plot(df1["date"], df1["soil"])
```

```python
plt.plot(df1["date"], df1["gwl"])
```
