```python
def remove_duplicate_date(data):
    return data[~data["date"].duplicated(keep='last')]

def convert_zero_to_na(data, column):
    data[column] = data[column].replace(0, np.nan)
    return data
```

```python
karang_agung = df[df["stasiun"] == "Karang agung"]
karang_agung = karang_agung.drop(["stasiun"], axis=1)
karang_agung = remove_duplicate_date(karang_agung)
karang_agung = karang_agung.set_index('date')
```

```python
karang_agung["gwl"] = karang_agung["gwl"] - 2.2
karang_agung
```

```python
plt.plot(karang_agung['gwl'])
```

```python
data = karang_agung
```

```python
data.loc[data['gwl'] <= -2.0, 'gwl'] = np.nan
data.loc[data['soil'] <= 5, 'soil'] = np.nan
data.loc[data['temp'] <= 20, 'temp'] = np.nan
```
