# Read Data

```python
df = pd.read_table("data/provinsi_sumsel (1).txt", delimiter=";",
                  names=['code', 'machine', 'provinsi', 'kabupaten', 'kecamatan', 'location', 'date', 'rain', 'soil', 'gwl', 'temp'],
                  header=None)
df
```

```python
df.dtypes
```

# Change data type

```python
df['rain'] = pd.to_numeric(df['rain'], errors='coerce')
df['soil'] = pd.to_numeric(df['soil'], errors='coerce')
df['gwl'] = pd.to_numeric(df['gwl'], errors='coerce')
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
```
